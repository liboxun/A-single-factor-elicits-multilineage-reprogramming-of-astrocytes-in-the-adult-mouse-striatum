```python
adata_neural = sc.read(neural_data_dir + 'ReprAlone_NeuralOnly_postFiltering_log_pcs50_Harmony_louvain_ann_rankPseudotime.h5ad')
```

```python
adata_neural = adata_neural[adata_neural.obs['library_name'] == 'BL41_reprogrammed']
```

```python
cell_proportions = adata_neural.obs['louvain'].value_counts()/adata_neural.shape[0]
```

```python
cell_proportions
```

```python
astrocyte_prop = cell_proportions.loc[['Astrocytes-1', 'Astrocytes-2', 'Astrocytes-3']].sum()
```

```python
astrocyte_prop
```

```python
L_astrocyte_prop = cell_proportions.loc[['Lenti-astrocytes']].sum()
```

```python
L_astrocyte_prop
```

```python
NPC_prop = cell_proportions.loc[['NPCs-1', 'NPCs-2']].sum()
```

```python
NPC_prop
```

```python
NB_prop = cell_proportions.loc[['Neuroblasts-1', 'Neuroblasts-2', 'Neuroblasts-3', 'Neuroblasts-4', 'Neuroblasts-5']].sum()
```

```python
NB_prop
```

```python
neuron_prop = cell_proportions.loc[['Neurons']].sum()
```

```python
neuron_prop
```

```python
palette = sns.color_palette("tab10")
```

```python
# Plot cell type compoisition of BL41 and BL42
fig, ax = plt.subplots(figsize=(5,1))
N = 1

ind = np.arange(N)    # the x locations for the groups
width = 0.1       # the width of the bars: can also be len(x) sequence


ax.barh(ind, astrocyte_prop, align='center', color=palette[0])
ax.barh(ind, L_astrocyte_prop, align='center', 
        left=astrocyte_prop, color=palette[1])
ax.barh(ind, NPC_prop, align='center', 
        left=astrocyte_prop+L_astrocyte_prop, color=palette[2])
ax.barh(ind, NB_prop, align='center', 
        left=astrocyte_prop+L_astrocyte_prop+NPC_prop, color=palette[3])
ax.barh(ind, neuron_prop, align='center', 
        left=astrocyte_prop+L_astrocyte_prop+NPC_prop+NB_prop, color=palette[4])


ax.set_yticks(ind)
ax.invert_yaxis()  # labels read top-to-bottom
# ax.set_xlabel('Performance')
# ax.set_title('How fast do you want to go today?')
# plt.legend(('BL41_reprogrammed', 'BL42_control'))

plt.savefig('./figures/hbar_cell_type_composition.pdf')

plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Astrocytes', 'Lenti-astrocytes', 'NPCs', 'Neuroblasts', 'Neurons'
 #sizes = [15, 30, 45, 10]
sizes = [astrocyte_prop, L_astrocyte_prop, NPC_prop, NB_prop, neuron_prop]
explode = (0.02, 0.02, 0.02, 0.02, 0.02)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, 
        #autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.savefig('./figures/pie_cell_type_composition.pdf')


plt.show()
```
